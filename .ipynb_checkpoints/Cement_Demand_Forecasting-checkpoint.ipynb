## KPI analysis on stockout, overstocking, reactive ordering, and poor visibility



```python
#Calculating the KPI analysis on stockout 

df['stockout'] = df['planned_pour_tonnes'] > (df['opening_inventory_tonnes']) + df['deliveries_tonnes']
```

```python
#Calculating the KPI analysis on overstocking

df['overstocking'] = df['closing_inventory_tonnes'] > df['silo_capacity']
```

```python
#Calculating the KPI analysis on idle

df['idle'] = (df['planned_pour_tonnes'] == 0) & (df['deliveries_tonnes']== 0)
```

```python
#Calculating the KPI analysis on waste risk
df['waste_risk'] = (df['closing_inventory_tonnes'] > 0.85 * df['silo_capacity']) & (df['planned_pour_tonnes'] < 5)
```

```python
#Calculating the KPI analysis on pour_disrupt
df['pour_disrupt'] = (df['planned_pour_tonnes'] > 0) & (df['consumed_tonnes'] == 0)
```

***To get the KPI summary***

```python
kpi_summary = df.groupby('site_id').agg(
    total_days=('date','count'),
    total_consumed_tonnes=('consumed_tonnes', 'sum'),
    avg_daily_consumed=('consumed_tonnes', 'mean'),
    stockout_pct=('stockout', lambda x: round(x.mean() * 100,2)),
    overstocking_pct=('overstocking', lambda x: round(x.mean() * 100,2)),
    idle_pct=('idle', lambda x: round(x.mean() * 100,2)),
    waste_risk_pct=('waste_risk', lambda x: round(x.mean() * 100,2)),
    pour_disrupt_pct=('pour_disrupt', lambda x: round(x.mean() * 100,2)),
    silo_capacity=('silo_capacity', 'first'),
    region=('region', 'first')
).reset_index()
```

```python
kpi_summary
```

```python
kpi_summary.to_csv('historical_kpi_summary.csv', index=False)
```

```python
# Cell 2: Comprehensive KPI Calculation

def calculate_comprehensive_kpis(df):
    df['pour_ready'] = (df['opening_inventory_tonnes'] + df['deliveries_tonnes']) >= df['planned_pour_tonnes']
    df['silo_utilization'] = df['closing_inventory_tonnes'] / df['silo_capacity']
    df['write_off_risk'] = (df['closing_inventory_tonnes'] > 0.9 * df['silo_capacity']) & (df['consumed_tonnes'] < 10)
    
    return df.groupby('site_id').agg({
        'pour_ready': lambda x: round(x.mean() * 100, 2),
        'silo_utilization': lambda x: round(x.mean() * 100, 2),
        'write_off_risk': lambda x: round(x.mean() * 100, 2),
        'consumed_tonnes': 'sum',
        'silo_capacity': 'first'
    }).reset_index()

comprehensive_kpis = calculate_comprehensive_kpis(df)
print(f"Average Pour Readiness: {comprehensive_kpis['pour_ready'].mean():.1f}%")
print(f"Average Silo Utilization: {comprehensive_kpis['silo_utilization'].mean():.1f}%")
print(f"Average Write-off Risk: {comprehensive_kpis['write_off_risk'].mean():.1f}%")
```

***ARIMA and DECISION_TREE***

```python
from statsmodels.tsa.arima.model import ARIMA
import numpy as np
```

```python
import warnings 
warnings.filterwarnings('ignore')
```

```python
#Creating a subset from the site_id before considering the baseline for this model testing 
#creating a dataframe from the existing one 

site_id = 'SITE_001'
site_df = df[df['site_id'] == site_id].copy()
```

```python
site_df.set_index('date', inplace = True)
site_df = site_df.sort_index()
```

```python
site_df.head()
```

```python
#Creating a baseline that will be use to train and test the model 

y = site_df['consumed_tonnes']

exog = site_df[['planned_pour_tonnes','rain_mm','avg_temp_c']]
```

```python
#Creating a split index to train and test the baseline

split_index = int(len(site_df) * 0.8) #considering 80% of the dataframe
X_train, X_test = exog.iloc[:split_index], exog.iloc[split_index:]
Y_train, Y_test = y.iloc[:split_index], y.iloc[split_index:]
```

```python
#Predicting the trained model

Arima_forecast = result.predict(start=len(Y_train), 
                           end=len(Y_train) + len(Y_test) - 1,
                           exog=X_test)
```

***Evaluating the ARIMA model to check if MAPE is  15%***

```python
from sklearn.metrics import mean_squared_error
```

```python
def print_metric(y_true, y_pred, label):
    mask = y_true != 0
    mape = np.mean(np.abs((y_true[mask] - y_pred[mask]) / y_true[mask])) * 100
    rmse = np.sqrt(mean_squared_error(y_true, y_pred))
    print(f'{label} - MAPE: {mape:.2f}%, RMSE: {rmse:.2f}tonnes')
```

```python
print_metric(Y_test, Arima_forecast, label="ARIMA Model")
```

***Applying the Machine-Learning Model (Decision tree) to further Train and evaluate the model***

```python
#Using the Decision Tree algorithm to train and evaluate the data

from sklearn.tree import DecisionTreeRegressor

dt_model = DecisionTreeRegressor(random_state=42)
dt_model.fit(X_train, Y_train)
```

```python
#Predicting the model
dt_forecast = dt_model.predict(X_test)
```

# Feature Engineering

***Adding more features to the datasets, to enable the model run better and generate a good result***

```python
#Creating Lag Features

df['lag_1'] = df.groupby('site_id')['consumed_tonnes'].shift(1)
df['lag_3'] = df.groupby('site_id')['consumed_tonnes'].shift(3)
df['lag_7'] = df.groupby('site_id')['consumed_tonnes'].shift(7)

#Creating Rolling Window Features

df['rolling_mean_3'] = df.groupby('site_id')['consumed_tonnes'].rolling(3).mean().reset_index(level=0, drop= True)
df['rolling_std_7'] = df.groupby('site_id')['consumed_tonnes'].rolling(7).std().reset_index(level=0, drop= True)

#Creating Time-Based Features

df['day_of_week'] = df['date'].dt.dayofweek   # Monday=0, Sunday=6
df['week_of_year'] = df['date'].dt.isocalendar().week
df['month'] = df['date'].dt.month

#Interaction Features

df['rain_x_pour'] = df['rain_mm'] * df['planned_pour_tonnes']
df['temp_x_pour'] = df['avg_temp_c'] * df['planned_pour_tonnes']

#Encoding Behavior 

df['behavior_encoded'] = df['behavior'].astype('category').cat.codes
df['cement_type_encoded'] = df['cement_type'].astype('category').cat.codes

#Adding site avergae 

site_avg = df.groupby('site_id')['consumed_tonnes'].transform('mean')
df['site_id_consumption'] = site_avg
```

***Predicting cement demand up to 8 weeks ahead***

```python

def generate_8week_forecast(df, site_id):
    site_df = df[df['site_id'] == site_id].copy().sort_values('date')
    site_df.set_index('date', inplace=True)

    # --- Create lag & calendar features ---
    site_df['lag_1'] = site_df['consumed_tonnes'].shift(1)
    site_df['rolling_mean_7'] = site_df['consumed_tonnes'].rolling(7).mean()
    site_df['month'] = site_df.index.month
    site_df['day_of_week'] = site_df.index.dayofweek

    # --- Work out which feature columns actually exist in this dataframe ---
    possible_features = [
        'planned_pour_tonnes',
        'rain_mm',
        'avg_temp_c',
        'lag_1',
        'rolling_mean_7',
        'month',
        'day_of_week'
    ]
    features = [col for col in possible_features if col in site_df.columns]

    if len(features) == 0:
        raise ValueError("No valid feature columns found in the dataframe for this site.")

    # Drop rows with missing values in used features or target
    site_df = site_df.dropna(subset=features + ['consumed_tonnes'])

    if site_df.empty:
        raise ValueError(f"No data available for site {site_id} after preprocessing.")

    X = site_df[features]
    y = site_df['consumed_tonnes']

    model = DecisionTreeRegressor(random_state=42)
    model.fit(X, y)

    # --- Build future index ---
    last_date = site_df.index[-1]
    future_dates = pd.date_range(start=last_date + pd.Timedelta(days=1), periods=56, freq='D')
    future_df = pd.DataFrame(index=future_dates)

    # --- Only create future feature columns that were actually using ---
    if 'planned_pour_tonnes' in features:
        future_df['planned_pour_tonnes'] = site_df['planned_pour_tonnes'].mean()

    if 'rain_mm' in features:
        future_df['rain_mm'] = site_df['rain_mm'].mean()
        # If your column is actually 'rain_forecast_mm', you could do:
        # future_df['rain_mm'] = site_df['rain_forecast_mm'].mean()

    if 'avg_temp_c' in features:
        future_df['avg_temp_c'] = site_df['avg_temp_c'].mean()

    if 'month' in features:
        future_df['month'] = future_df.index.month

    if 'day_of_week' in features:
        future_df['day_of_week'] = future_df.index.dayofweek

    if 'lag_1' in features:
        future_df['lag_1'] = site_df['consumed_tonnes'].iloc[-1]

    if 'rolling_mean_7' in features:
        future_df['rolling_mean_7'] = site_df['consumed_tonnes'].tail(7).mean()

    forecasts = []

    for i in range(56):
        # Predict for the current day
        X_future = future_df.iloc[i:i+1][features]
        pred = model.predict(X_future)[0]
        forecasts.append(pred)

        # Update lag_1 and rolling_mean_7 for future rows if used
        if i < 55:
            if 'lag_1' in features:
                future_df.iloc[i+1, future_df.columns.get_loc('lag_1')] = pred

            if 'rolling_mean_7' in features:
                if i < 6:
                    historical_values = site_df['consumed_tonnes'].tail(7 - i - 1).tolist()
                    new_values = forecasts
                    rolling_values = historical_values + new_values
                    future_df.iloc[i+1, future_df.columns.get_loc('rolling_mean_7')] = sum(rolling_values) / len(rolling_values)
                else:
                    future_df.iloc[i+1, future_df.columns.get_loc('rolling_mean_7')] = sum(forecasts[i-6:i+1]) / 7

    return pd.DataFrame({
        'date': future_dates,
        'site_id': site_id,
        'forecasted_consumption': forecasts,
        'week': [(i // 7) + 1 for i in range(56)]
    })

# Example usage:
forecast_8week = generate_8week_forecast(df, 'SITE_001')
print("8-Week Forecast Summary by Week:")
print(forecast_8week.groupby('week')['forecasted_consumption'].sum().round(2))
```

## Inventory Simulation

## Scaling Pipeline

```python
#Defining the feature engineers used for site_id

def engineer_features(df, site_id):
    site_df = df[df['site_id'] == site_id].copy().sort_values('date')
    site_df.set_index('date', inplace=True)

    site_df['lag_1'] = site_df['consumed_tonnes'].shift(1)
    site_df['lag_3'] = site_df['consumed_tonnes'].shift(3)
    site_df['lag_7'] = site_df['consumed_tonnes'].shift(7)

    site_df['rolling_mean_3'] = site_df['consumed_tonnes'].rolling(3).mean()
    site_df['rolling_std_7'] = site_df['consumed_tonnes'].rolling(7).std()


    site_df['rain_x_pour'] = site_df['rain_mm'] * site_df['planned_pour_tonnes']
    site_df['temp_x_pour'] = site_df['avg_temp_c'] * site_df['planned_pour_tonnes']

    site_df['inventory_gap'] = site_df['opening_inventory_tonnes'] + site_df['deliveries_tonnes'] - site_df['planned_pour_tonnes']
    site_df['inventory_ratio'] = site_df['closing_inventory_tonnes'] / site_df['silo_capacity']

    site_df.dropna(inplace=True)
    return site_df
```

```python
def train_dt_forecast(site_df):
    features = [
        'planned_pour_tonnes','rain_mm', 'avg_temp_c', 'rolling_mean_3',
       'rolling_std_7', 'week_of_year', 'rain_x_pour',
       'temp_x_pour', 'inventory_gap', 'inventory_ratio'
    ]

    X = site_df[features]
    y = site_df['consumed_tonnes']

    split_index = int(len(site_df) * 0.8)
    X_train, X_test = X.iloc[:split_index], X.iloc[split_index:]
    y_train, y_test = y.iloc[:split_index], y.iloc[split_index:]

    dt_model2 = DecisionTreeRegressor(random_state=42)
    dt_model2.fit(X_train, y_train)
    dt_forecast2 = dt_model2.predict(X_test)

    test_results = site_df.iloc[split_index:].copy()
    test_results['forecasted_consumption'] = dt_forecast2

    return dt_model2, test_results
```

```python
def simulate_inventory(test_results, site_meta, lead_time_days=2, buffer_rain_threshold=10, buffer_increase=0.1):
    df_sim = test_results[['consumed_tonnes', 'forecasted_consumption', 'rain_mm']].copy()
    df_sim['sim_inventory'] = np.nan
    df_sim['reorder_flag'] = False
    df_sim['recommended_delivery_date'] = None
    df_sim['recommended_delivery_quantity'] = 0.8
    df_sim['buffer_applied'] = False

    silo_capacity = site_meta['silo_capacity']
    inventory = site_meta['initial_inventory']

    reorder_threshold = site_meta['reorder_threshold']
    target_inventory = site_meta['target_inventory']

    delivery_queue = {}

    for today, row in df_sim.iterrows():
        if today in delivery_queue:
            inventory += delivery_queue[today]
            inventory = min(inventory, silo_capacity)
            del delivery_queue[today]

        consumption = row['forecasted_consumption']
        inventory -= consumption

        df_sim.loc[today, 'sim_inventory'] = inventory

        if inventory < reorder_threshold:
            df_sim.loc[today, 'reorder_flag'] = True
            delivery_date = today + pd.Timedelta(days=lead_time_days)

            delivery_qty = target_inventory - inventory
            if row['rain_mm'] > buffer_rain_threshold:
                delivery_qty *= (1 + buffer_increase)
                df_sim.loc[today, 'buffer_applied'] = True

            delivery_qty = min(delivery_qty, silo_capacity - inventory)
            delivery_queue[delivery_date] = delivery_qty

            df_sim.loc[today, 'recommended_delivery_date'] = delivery_date.strftime('%Y-%m-%d')
            df_sim.loc[today, 'recommended_delivery_quantity'] = round(delivery_qty, 2)

    return df_sim
```

```python
def run_pipeline(df, site_metedata):
    all_results = []

    for site_id in df['site_id'].unique():
        print(f'Processing {site_id}.....')
        site_df = engineer_features(df, site_id)

        if len(site_df) < 50:
            print(f'Skipping {site_id}, not enough data or no data at all')
            continue

        dt_model2, test_results = train_dt_forecast(site_df)
        sim_results = simulate_inventory(test_results, site_metadata[site_id])
        sim_results['site_id'] = site_id

        all_results.append(sim_results.reset_index())

    return pd.concat(all_results, ignore_index=True)
```

```python
for site_id in df['site_id'].unique():
    print(f'Evaluating model for {site_id}.....')

    site_df = engineer_features(df, site_id)

    if len(site_df) < 50:
        print(f'skipping for {site_id}, not enough data')

    features = [
        'planned_pour_tonnes', 'rain_mm', 'avg_temp_c',
        'lag_1', 'lag_3', 'lag_7',
        'rolling_mean_3', 'rolling_std_7',
        'rain_x_pour', 'temp_x_pour',
        'inventory_gap', 'inventory_ratio'
    ]

    X = site_df[features]
    y = site_df['consumed_tonnes']

    split_index = int(len(site_df) * 0.8)
    X_train, X_test = X.iloc[:split_index], X.iloc[split_index:]
    y_train, y_test = y.iloc[:split_index], y.iloc[split_index:]

    dt_model2 = DecisionTreeRegressor(random_state=42)
    dt_model2.fit(X_train, y_train)
    
    dt_forecast2 = dt_model2.predict(X_test)

    mask = y_test != 0
    if mask.sum() > 0:
        mape = mean_absolute_percentage_error(y_test[mask], dt_forecast2[mask]) * 100
    else:
        mape = 0

    rmse = np.sqrt(mean_squared_error(y_test, dt_forecast2))

    performance_summary.append(
        {
            'site_id': site_id,
            'test_samples': len(y_test),
            'mape': mape,
            'rmse': rmse,
            'avg_consumption': y_test.mean()
        }
    )
```
